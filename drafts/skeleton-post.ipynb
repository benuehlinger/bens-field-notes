```python
rng = np.random.default_rng(42)
n = 5000

df = pd.DataFrame(
    {
        "fico": rng.normal(720, 75, n).clip(500, 850).astype(int),
        "balance": rng.lognormal(np.log(22_000), 0.45, n),
    }
)

df.head()
```
